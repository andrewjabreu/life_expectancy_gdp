```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
life_expectency = pd.read_csv('all_data.csv')
print(life_expectency.head())
```

```python
print(life_expectency.info())
print(life_expectency.describe())
print(life_expectency['Country'].unique())
```

```python
chile = life_expectency[life_expectency.Country == 'Chile']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=chile)
plt.title('Chile GDP to Lifespan')
plt.ylabel('Gross Domestic Product')
plt.show()
plt.clf()
```

```python
China = life_expectency[life_expectency.Country == 'China']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=China,color='red')
plt.title('China GDP to Lifespan')
plt.ylabel('Gross Domestic Product')
plt.show()
plt.clf()
```

```python
Germany = life_expectency[life_expectency.Country == 'Germany']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=Germany,color='green')
plt.title('Germany GDP to Lifespan')
plt.ylabel('Gross Domestic Product')
plt.show()
plt.clf()
```

```python
Mexico = life_expectency[life_expectency.Country == 'Mexico']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=Mexico,color='purple')
plt.title('Mexico GDP to Lifespan')
plt.ylabel('Gross Domestic Product')
plt.show()
plt.clf()
```

```python
USA = life_expectency[life_expectency.Country == 'United States of America']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=USA,color='m')
plt.title('USA GDP to Lifespan')
plt.ylabel('Gross Domestic Product')
plt.show()
plt.clf()
```

```python
Zimbabwe = life_expectency[life_expectency.Country == 'Zimbabwe']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=Zimbabwe,color='salmon')
plt.title('Zimbabwe GDP to Lifespan')
plt.ylabel('Gross Domestic Product')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(15,9))

plt.subplot(2,3,1)
chile = life_expectency[life_expectency.Country == 'Chile']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=chile)
plt.title('Chile')
plt.xlabel('Life expectancy (years)')
plt.ylabel('Gross Domestic Product')

plt.subplot(2,3,2)
China = life_expectency[life_expectency.Country == 'China']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=China,color='red')
plt.title('China')
plt.xlabel('Life expectancy (years)')
plt.ylabel('Gross Domestic Product')


plt.subplot(2,3,3)
Germany = life_expectency[life_expectency.Country == 'Germany']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=Germany,color='green')
plt.title('Germany')
plt.xlabel('Life expectancy (years)')
plt.ylabel('Gross Domestic Product')

plt.subplot(2,3,4)
Mexico = life_expectency[life_expectency.Country == 'Mexico']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=Mexico,color='blueviolet')
plt.title('Mexico')
plt.xlabel('Life expectancy (years)')
plt.ylabel('Gross Domestic Product')

plt.subplot(2,3,5)
USA = life_expectency[life_expectency.Country == 'United States of America']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=USA,color='magenta')
plt.title('USA')
plt.xlabel('Life expectancy (years)')
plt.ylabel('Gross Domestic Product')

plt.subplot(2,3,6)
Zimbabwe = life_expectency[life_expectency.Country == 'Zimbabwe']
sns.scatterplot(x='Life expectancy at birth (years)',y='GDP',data=Zimbabwe,color='salmon')
plt.title('Zimbabwe')
plt.xlabel('Life expectancy (years)')
plt.ylabel('Gross Domestic Product')

plt.subplots_adjust(wspace=.5,hspace=.5)

legend_labels = ['Chile','China','Germany','Mexico','USA','Zimbabwe']

plt.figlegend(labels=legend_labels)
plt.show()
plt.clf()
```

```python
# The USA had the most visually consistent association between GDP and Life expectancy
```
